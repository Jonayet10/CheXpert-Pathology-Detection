### Local Binary Pattern (LBP)

https://scikit-image.org/docs/stable/auto_examples/features_detection/plot_local_binary_pattern.html 

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import feature, io
import os

# lbp params
R = 4 # radius of circle (spatial resolution of operator)
P = 8  # num points to be considered neighbors
METHOD = "uniform"

# image params
img_dir = "image_samples"
```

```python
# calculate lbp
def local_binary_pattern(image, P, R):
    lbp = feature.local_binary_pattern(image, P, R, method=METHOD)
    return lbp

img_lst = []

# image list contains all of the image paths
for filename in os.listdir(img_dir):
    full_path = os.path.join(img_dir, filename)  # Create the full path to the file
    img_lst.append(full_path)  # Add it to the list
```

```python
def display_image(img_lst):
    for image_src in img_lst: 
        image = io.imread(image_src)
        lbp_image = local_binary_pattern(image, P, R)

        # show images
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
        ax1.imshow(image, cmap='gray')
        ax1.set_title('Original Image')
        ax1.axis('off')
        ax2.imshow(lbp_image, cmap='gray')
        ax2.set_title('LBP Image')
        ax2.axis('off')

        plt.show()

display_image(img_lst)
```
