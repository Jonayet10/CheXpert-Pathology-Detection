### Sobel Filter

https://scikit-image.org/docs/stable/api/skimage.filters.html#skimage.filters.sobel

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, filters
import os

# image params
img_dir = "image_samples"
```

```python
img_lst = []

# image list contains all of the image paths
for filename in os.listdir(img_dir):
    full_path = os.path.join(img_dir, filename)  # Create the full path to the file
    img_lst.append(full_path)  # Add it to the list

img_lst
```

```python
def display_image(img_lst):
    for img in img_lst: 
        image = io.imread(img)
        sobel = filters.sobel(image)  

        fig, axes = plt.subplots(1, 2, figsize=(16, 4))
        ax = axes.ravel()

        ax[0].imshow(image, cmap='gray')
        ax[0].set_title('Original Image')
        ax[1].imshow(sobel, cmap='gray')
        ax[1].set_title('Sobel Magnitude')

        for a in ax:
            a.axis('off')

        plt.tight_layout()
        plt.show()

display_image(img_lst)
```
